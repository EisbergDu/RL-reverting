```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

def plot_cliff_walking(ncol=12, nrow=4):
    fig, ax = plt.subplots(figsize=(10, 4))
    
    # 绘制网格线
    ax.set_xticks(np.arange(0, ncol+1, 1))
    ax.set_yticks(np.arange(0, nrow+1, 1))
    ax.grid(True, linewidth=1, color='black')
    ax.set_xticklabels([])
    ax.set_yticklabels([])
    
    # 设置坐标轴范围
    ax.set_xlim(0, ncol)
    ax.set_ylim(0, nrow)
    
    # 起点：左下角 (0, 0) 但在我们的坐标中列索引0，行索引3（最后一行）
    start_col, start_row = 0, nrow-1
    # 终点：右下角 (ncol-1, nrow-1)
    goal_col, goal_row = ncol-1, nrow-1
    
    # 悬崖区域：底部行，从第1列到第ncol-2列
    cliff_cols = list(range(1, ncol-1))
    
    # 着色悬崖
    for col in cliff_cols:
        rect = patches.Rectangle((col, nrow-1), 1, 1, linewidth=0, facecolor='gray', alpha=0.5)
        ax.add_patch(rect)
    
    # 标记起点 (S)
    ax.text(start_col + 0.5, start_row + 0.5, 'S', ha='center', va='center', fontsize=20, fontweight='bold', color='green')
    
    # 标记终点 (G)
    ax.text(goal_col + 0.5, goal_row + 0.5, 'G', ha='center', va='center', fontsize=20, fontweight='bold', color='red')
    
    # 标记悬崖字样（可选择）
    for col in cliff_cols:
        ax.text(col + 0.5, nrow-1 + 0.5, 'X', ha='center', va='center', fontsize=14, color='black', alpha=0.7)
    
    # 反转y轴使第一行在顶部（可选，但习惯上网格第一行在上方）
    ax.invert_yaxis()
    
    # 添加标题和说明
    ax.set_title('Cliff Walking Environment', fontsize=16)
    ax.annotate('Green S = Start\nRed G = Goal\nGray area = Cliff (reward -100, terminal)', 
                xy=(0.5, -0.2), xycoords='axes fraction', ha='center', fontsize=10)
    
    plt.tight_layout()
    plt.show()

plot_cliff_walking()
```
